#### Iterators

```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
## Iterator
iterator=iter(my_list)
print(type(iterator))
```

```python
iterator
```

```python
## iterate through all the elements
next(iterator)
```

```python

try:
    print(next(iterator))
except StopIteration:
    print("There are no elements in the iterator")
```
