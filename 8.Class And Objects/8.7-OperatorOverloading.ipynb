```python
## common operators overloading magic methods
'''
__add__(self,other):Adds two obj using + operator.
__sub__(self,other):subtracts two obj using the - operator.
__mul__(self,other):multiplies two obj using the * operator.
__truediv__(self,other):divides two objects using the / operator.
__eq__(self,other):checks if two obj are equal using the == operator.
__lt__(self,other):checks if one obj is less than another using the < operator.
__gt__
'''
```

```python
class Vector:
    def __init__(self,x,y):
        self.x=x
        self.y=y

    def __add__(self,other):
        return Vector(self.x+other.x,self.y+other.y)

    def __sub__(self,other):
        return Vector(self.x-other.x,self.y-other.y)

    def __mul__(self,other):
        return Vector(self.x*other.x,self.y*other.y)
    
    def __eq__(self,other):
        return Vector(self.x==other.x,self.y==other.y)

    def __repr__(self):
        return f"Vector({self.x},{self.y})"

## create obj of the vector class
v1=Vector(2,3)
v2=Vector(4,5)

print(v1+v2)
print(v1-v2)
    
```
