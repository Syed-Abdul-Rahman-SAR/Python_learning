```python
## tuple methods
print(numbers)
print(numbers.count(1))#number of times 1 appear
print(numbers.index(3))##index of 3
```

```python
## packing and unpacking tuple
## packing
packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
##unpacking
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
##Unpaking with *
numbers=(1,2,3,4,5,6)
first,second,*middle,last=numbers
print(first)
print(second)
print(middle)
print(last)
```

```python
##Nested Tuple
##Nested List
lst=[[1,2,3,4],[5,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python
lst=[[1,2,3,4],[5,7,8,9],(1,"Hello",3.14,"c")]
lst[2][0:3]
```

```python
nested_tuple=((1,2,3),("a","b","c"),(True,False))

print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## iterating over nested tuple
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
