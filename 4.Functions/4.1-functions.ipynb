```python
## default parameters

def greet(name):
    print(f"Hello {name} Welcome")
greet("Abdul")
```

```python
def greet(name="Guest"):
    print(f"Hello {name} Welcome")

greet("Abdul")
greet()
```

```python
## variable length arguments 
## positional and keyword arguments
## positional arguments

def print_number(*args):
    for number in args:
        print(number)
```

```python
print_number(1,2,3,4,5,6,"Abdul")
```

```python
## keywords Argumenst

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Abdul",age="32",country="india")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"positional arguments {val}")
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,name="Abdul",age="32",country="India")
```
